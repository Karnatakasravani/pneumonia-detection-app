```python
arr = [1,4,2,5,6,7,8]
k = 7
summ = 0
cnt  =0 
d={}
for i in arr:
    summ+= i
    if summ - k  in d:
        cnt+= d[summ-k]
    if summ not in d:
        d[summ] = 1
    else:
        d[summ]+=1
print(cnt)
        
```

```python
string = "banana"
max_len = 1
max_str = string[0]
for i in range(len(string)):
    for j in range(i+1,len(string)):
        if j-i+1>max_len and string[i:j+1] == string[i:j+1][::-1]:
            max_str = string[i:j+1]
            max_len = j-i+1
print(max_str)
```

```python
def unique_string(arr):
    unique_str = set()
    max_len = 0
    left = 0
    for right in range(len(arr)):
        while arr[right] in unique_str:
            unique_str.remove(arr[left])
            left += 1
        unique_str.add(arr[right])
        max_len = max(max_len , right - left +1)
    return max_len
n = "banana"
print(unique_string(n))
```

```python
def steps(n):
    dp = [0] * (n + 1)
    dp[0] = 1  
    dp[1] = 1  
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    
    return dp[n]

n = int(input())  # Input the number of stairs
print(steps(n))   # Output the number of ways to reach the nth stair
```
